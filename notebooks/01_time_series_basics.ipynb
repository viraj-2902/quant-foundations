```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf
```

```python
spy = yf.download("SPY", start="2015-01-01")

type(spy.index)

spy.columns

spy.xs("SPY", level="Ticker",axis=1)

```

```python
close_spy = spy["Close"]


log_price = np.log(close_spy)

log_price_lag = log_price.shift(1)

log_returns = log_price - log_price_lag

```

```python
plt.plot(close_spy)
plt.title("Closing Price of SPY from 2015-2025")
plt.xlabel("Years")
plt.ylabel("Closing Price Each Day")
```

```python
plt.plot(log_returns)
plt.title("Returns of SPY from 2015 - 2025")
plt.xlabel("Years")
plt.ylabel("Returns")
```
